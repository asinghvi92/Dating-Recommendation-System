```python
from collections import Counter
import numpy as np
import json
import time
import heapq
import ast
import sys

# command: python compatibility_score.py infile outfile_nonML outfile_ML max_users
#from features import compatibilityScoreCompute
#from MLtrainingtesting import MLLabelAssignment

t0 = time.time()
#users_json = []

#args = sys.argv
#infile = open(args[1], 'r')
#outfile = open(args[2], 'w')
#outfile_ml = open(args[3], 'w')
#max_users = args[4]

infile = open('../dataset/users_evaluation.json', 'r')
outfile = open('output/compatibility.json', 'w')
outfile_ml = open('output/compatibility_ml.json', 'w')
max_users = 1000

top_k = 5
#max_users = 100

users_json = ast.literal_eval(infile.readline())
length = len(users_json)

print "Done reading users file. Number of users found = ", length, "!"

t1 = time.time()
print "Creating/Initializing Score matrix... Time: ", t1-t0
S = np.zeros((max_users, max_users), dtype=np.float)
S_ml = np.zeros((max_users, max_users), dtype=np.float)

for x in range(0, max_users):
    S[x][x] = -1.0
    S_ml[x][x] = -1.0

t2 = time.time()
print 'Created and initialized Score matrix of size ', len(S), 'x', len(S[0]), ' - Time: ', t2-t1

print 'Extracting random ', max_users ,'users...'

users = []
i = 0
for user in users_json:
    if (i >= max_users):
        break
    users.append(user)
    i += 1
#print 'users: ', users
print 'Computing Compatibility Scores...'
keys = {}
index_X = 0

for key_X in users:
    keys[index_X] = key_X  
    index_Y = 0
    
    if (index_X == max_users):
        break
    
    for key_Y in users: 
        if (index_X != index_Y):
            if (index_Y > index_X):  
                score = 0.0
                score_ml = 0.0
                if (orientation_match(users_json[key_X], users_json[key_Y]) and language_match(users_json[key_X], users_json[key_Y])):
                    score = compatibilityScoreCompute(users_json[key_X], users_json[key_Y])
                    score_ml = MLLabelAssignment(users_json[key_X], users_json[key_Y])
                else:
                    score = 0.0
                    score_ml = 0.0
                
                
                #score = compability_score(users_json[key_X], users_json[key_Y])
                #print 'index_X', index_X, 'index_Y', index_Y, 'score', score
                S[index_X][index_Y] = score
                S[index_Y][index_X] = score
                
                S_ml[index_X][index_Y] = score_ml
                S_ml[index_Y][index_X] = score_ml
        index_Y += 1
    index_X += 1

#print 'S = ', S
t3 = time.time()
print 'Done computing scores! Time: ', t3-t2

compatibility = {}
compatibility_ml = {}
print 'Extracting top ', top_k, ' users'
for i in range(max_users):
    indexes = heapq.nlargest(top_k, xrange(len(S[i])), S[i].__getitem__)
    cusers = []
    for j in indexes:
        cusers.append(keys[j])
    compatibility[keys[i]] = cusers
    
    indexes_ml = heapq.nlargest(top_k, xrange(len(S_ml[i])), S_ml[i].__getitem__)
    cusers_ml = []
    for j in indexes_ml:
        cusers_ml.append(keys[j])
    compatibility_ml[keys[i]] = cusers_ml


print 'Writing output to file: ', outfile
print 'Writing ML output to file: ', outfile_ml

json.dump(compatibility, outfile)
outfile.close()

json.dump(compatibility_ml, outfile_ml)
outfile_ml.close()

t4 = time.time()
print 'Done! Time: ', t4-t3
print 'Total time: ', t4-t0
```
